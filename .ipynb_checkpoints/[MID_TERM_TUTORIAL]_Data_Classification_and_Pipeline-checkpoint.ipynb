### Filling Null Values & Convert to Numerical Values
Since we have some of the data that is null, we figure out how to fill the nulls. There are some methods to fill the null values.

1. **backward/forward fill**

  ```python
  df.fillna(method='bfill') # For backward fill
  df.fillna(method='ffill') # For forward fill
  ```

2. **feature engineer**
  See the pattern in the data, see if there any column that can fill the other column, e.g. if there are column `age` and column `born`, you can fill the `age` column by substracting the current time date by the `born` column

````python
def check_null(df):
    num_cols = []
    for col in df:
        num_cols.append(df[col].isnull().sum())
    num_cols = np.array(num_cols)
    pct_null = 100*(num_cols/len(df))
    return pd.DataFrame({'Column':df.columns, 'Null(s)': num_cols, 'Percentage': pct_null}).set_index('Column')

check_null(df)
````

There are some of the columns that have null values, which include

1. `age`
2. `embarked`
3. `deck`
4. `embark_town`

First, we deal with the `age` column first. We can see from the whole dataset if we want to fill the null value of column `age`, we fill it with the mean of the age according to the `who`. The `who` column tells us what is gender identity of the passenger is. We have to check each of the unique value appears in the column `who`.

```python
df['who'].value_counts()
```

````python
df['who'].value_counts()
````

Now we fill the null values in column `age` with the mean of each gender identity age.

````python
man_age_mean = df[df['who'] == 'man']['age'].mean()
woman_age_mean = df[df['who'] == 'woman']['age'].mean()
child_age_mean = df[df['who'] == 'child']['age'].mean()
````

````python
df['age'] = np.where(df['who'] == 'man', man_age_mean, df['age'])
df['age'] = np.where(df['who'] == 'woman', woman_age_mean, df['age'])
df['age'] = np.where(df['who'] == 'child', child_age_mean, df['age'])
````

After we fill up the null values in column `age`, we see there are null values in `deck` that counts up to **77%** of null values compared to the whole dataset. We can easily drop the column or you can find the pattern in the other column to make up the `deck` column.

````python
df.drop('deck', axis=1, inplace=True) # axis = 1 means that the axis that will be dropped is column axis
````

#### Drop unneccessary column
In the dataset, there are some column that can be represented by other column. for example,

1. `sex` and `adult_male`

  can be represented by column `who`

2. `alone`

  can be represented by column `sibsp`, `parch`

3. `pclass`

  can be represented by column `class`

4. `embarked`

  can be represented by column `embark_town`

4. `alive`

  can be represented by column `survived`

So we drop unneccessary column(s) that can be represented by other column to reduce variance in dataset while modeling.

````python
df.drop(['sex', 'adult_male', 'alone', 'class', 'embark_town', 'alive'], axis=1, inplace=True)
````

In column `embarked` there are 2 null values, we can drop them or find the pattern in the other column to fill the null value in `embarked` column. If we drop the column, then we will miss a lot of information that might be important.

**For the sake of this tutorial i will be dropping the null records. (axis=0) which means we will be dropping the null records.**

````python
df.dropna(axis=0, inplace=True)
````

````python
df.info()
````

#### Convert to Numerical Values
After finish with preprocessing we will pass the data to machine learning model. To pass the dataset to machine learning model, we need to convert the remaining object data to numerical values. Categorical column in the dataset including:

1. `embarked`
2. `who`
3. `alive`

````python
embarked_labels, embarked_unique = pd.factorize(df['embarked'])
df['embarked'] = embarked_labels

who_labels, who_unique = pd.factorize(df['who'])
df['who'] = who_labels
````

### Feature Scaling
Feature scaling is a technique to standardize the independent features in a data in a fixed range or scale. There are types of feature scaling, which include:

1. **Standardization**

  **Standardization** is a scaling technique where the values are centered around the mean with a unit standard deviation. This means that the mean of the attribute becomes zero and the resultant distribution has a unit standard deviation.

  <img src='https://toptipbio.com/wp-content/uploads/2018/07/Z-score-formula.jpg'>


2. **Normalization**

  Normalization is a technique often applied as part of data preparation for machine learning. The goal of normalization is to change the values of numeric columns in the dataset to use a common scale, without distorting differences in the ranges of values or losing information.

  * Min Max Scaling
  * Mean Normalization
  * Max Absolute Scaling
  * Robust Scaling etc.


````python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler() # creating an instance of the class object
df[['age', 'fare']] = sc.fit_transform(df[['age', 'fare']])  #fit and transforming StandardScaler the dataframe
df[['age', 'fare']].head()
````

## Data Modeling

#### Split dataset to train set and test set

````python
X = df.drop('survived', axis=1) # Dataset features
y = df['survived'] # Target feature / label

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=True, stratify=y) # Split the dataset to train and test set

print(f'Length of X train is {len(X_train)}')
print(f'Length of X test is {len(X_test)}')
print(f'Length of label for training is {len(y_train)}')
print(f'Length of label for testing is {len(y_test)}')
````

#### **Logistic Regression**

<img src='https://miro.medium.com/v2/resize:fit:1400/1*44qV8LhNzE5hPnta2PaaHw.png' alt='Logistic Regression Illustration' width='75%'>

Calculation behind the model are not covered in this tutorial, however if you are curious with how the logistic regression algorithm works kindly check:

1. [Logistic Regression Overview](https://towardsdatascience.com/logistic-regression-detailed-overview-46c4da4303bc)
2. [Logistic Regression and Decision Boundary](https://towardsdatascience.com/logistic-regression-and-decision-boundary-eab6e00c1e8)

````python
log_reg = LogisticRegression() # Assign to any variable you want (make sure to make it intuitive, usually assigned to model variable → model = LogisticRegression()

log_reg.fit(X_train, y_train) # Fitting the training feature and label (X_train and y_train) to the model
y_pred = log_reg.predict(X_test) # Predict the testing feature (X_test)
````

The predicted feature has been assigned to `y_pred` variable, now we check the performance of the existing **logistic regression** model with **accuracy**, **precision**, **recall**, and **f1 score**

````python
print(f"""
Accuracy Score : {log_reg.score(X_test, y_test)}
Precision Score : {precision_score(y_test, y_pred)}
Recall Score : {recall_score(y_test, y_pred)}
F1-Score : {f1_score(y_test, y_pred)}
""")
````

Or you just simply call the `classification_report()` function and it will yield the model performance with variety of scoring method

````python
print(classification_report(y_test, y_pred)) # Compare the true label of testing label and predicted label
````
